```python
mapping = []
for index, row in table.iterrows():
    
    uuid = row['uuid']
    
    meta = json.load(open('./{}.json'.format(uuid)))
    
    subm_id = meta['ancestors'][0]['submission_id']
    
    mp = {
        'uuid': uuid,
        'subm_id': subm_id
    }
    mapping.append(mp)
```

```python
mapping = pd.DataFrame(mapping)
```

```python
mapping.to_csv('mapping.csv')
```
