```python
from glob import glob
from matplotlib import pyplot as plt
```

```python
from pydicom import dcmread
from pydicom.data import get_testdata_file
```

```python
def plot_dcm_file(fname):
    print(fname)
    ds = dcmread(fname)
    plt.imshow(ds.pixel_array, cmap=plt.cm.gray)
    plt.show()
```

```python
FNAME = './../other_data/lung_TCGA_LUSC/doiJNLP-TCGA-LUSC-01-30-2017/TCGA-LUSC/TCGA-34-5239/04-15-1993-PTCTSKTHPET  CT TUMOR-92693/2.000000-LOW DOSE C- CHEST-99239/1-001.dcm'
```

```python
plot_dcm_file(FNAME)
```
